```python
from abc import ABC , abstractmethod

class payment(ABC):
    
    def __init__(self,name,email,phone_no,amount):
        self.name = name
        self.email = email
        self.phone_no = phone_no
        self.amount = amount

    @abstractmethod
    def pay(self):
        pass

class UPI_payment(payment):
    
    def pay(self):
        upi = input("Enter your UPI ID : ")
        print("UPI payment is processing...")
        print("-------------------------------------------------------------")
        print("Name : ",self.name)
        print("Email : ",self.email)
        print("Phone no : ",self.phone_no)
        print("Amount to be paid : Rs.",self.amount)
        print("UPI ID : ",upi)
        print("Rs. ",self.amount," paid successfully by ",upi)

class card_payment(payment):

    def pay(self):
        card_no = input("Enter your UPI ID : ")
        print("Card payment is processing...")
        print("-------------------------------------------------------------")
        print("Name : ",self.name)
        print("Email : ",self.email)
        print("Phone no : ",self.phone_no)
        print("Amount to be paid : Rs.",self.amount)
        print("Card no : ",card_no)
        print("Rs. ",self.amount," paid successfully by ",card_no)

class net_banking_payment(payment):

    def pay(self):
        bank = input("Enter your Bank Name : ")
        print("NetBanking payment is processing...")
        print("-------------------------------------------------------------")
        print("Name : ",self.name)
        print("Email : ",self.email)
        print("Phone no : ",self.phone_no)
        print("Amount to be paid : Rs.",self.amount)
        print("Bank : ",bank)
        print("Rs. ",self.amount," paid successfully by ",bank)

def process(name,email,phone,amount):
    print("-------------------PAYEMENT GATEWAY SYSTEM-------------------")
    print("Choose Payement Method")
    print("1. UPI Payment")
    print("2. Card Payment")
    print("3. NetBanking Payment")

    print("-------------------------------------------------------------")
    choice = int(input("Enter your choice for payment : "))
    print("-------------------------------------------------------------")
    if choice == 1:
        payment = UPI_payment(name,email,phone,amount)
    elif choice == 2:
        payment = card_payment(name,email,phone,amount)
    elif choice == 3:
        payment = net_banking_payment(name,email,phone,amount)
    else:
        print("Please enter the one of the following choices.")   
    payment.pay()

process("Rannu","rannu@123",12345678876,50000)
```
